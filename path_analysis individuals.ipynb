```python
from IPython.display import clear_output
import time
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pickle
from matplotlib.animation import FuncAnimation
from PIL import Image

try:
    # Load result from file
    with open('result.pkl', 'rb') as f:
        result = pickle.load(f)
except FileNotFoundError:
    # Handle the case where the file doesn't exist
    print("File 'result.pkl' does not exist. Creating a new empty result dictionary.")
    result = {}

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

def update(frame):
    ax.view_init(20, frame)

input_player = input("Enter the name of player: ")

for key, value_list in result.items():
    if key == input_player:
        x = [v[0] for v in value_list]
        y = [v[1] for v in value_list]
        z = range(len(value_list))
        ax.scatter(x, y, z, label=key)
        # Connect consecutive points with lines
        for i in range(len(value_list) - 1):
            ax.plot([x[i], x[i+1]], [y[i], y[i+1]], [z[i], z[i+1]], color='gray', alpha=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Z')
ax.set_zlabel('Range')

ax.legend()

ani = FuncAnimation(fig, update, frames=range(0, 360, 3), interval=50)
ani.save('plot_rotation_individual_3d.gif', writer='pillow')
```
